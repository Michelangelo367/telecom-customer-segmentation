# K-Meeans Clustering

```python
#Selecting columns for clusterisation with k-means
selected_cols = ["age","credit_amount", "duration_in_month"]
cluster_data = df.loc[:,selected_cols]
```

```python
def distributions(df):
    fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize=(16,12))
    sns.distplot(df["age"], ax=ax1)
    sns.distplot(df["credit_amount"], ax=ax2)
    sns.distplot(df["duration_in_month"], ax=ax3)
    plt.tight_layout()
```

```python
distributions(cluster_data)
```

We can see that distributions are right-skewed. To obtain better results we should remove the skewness by logarithmic transformation. After that let's see how they look like.

```python
cluster_log = np.log(cluster_data)
distributions(cluster_log)
```

```python
scaler = StandardScaler()
cluster_scaled = scaler.fit_transform(cluster_log)
```

```python
clusters_range = range(2,14)
inertias =[]

for c in clusters_range:
    kmeans = KMeans(n_clusters=c, random_state=0).fit(cluster_scaled)
    inertias.append(kmeans.inertia_)

plt.figure()
plt.plot(clusters_range,inertias, marker='o')
```

The plot above shows that inertia decreases with increasing number of clusters.

This method allows for assessment of cluster separations and fitting of each observation in its own cluster. The highest score the better. I will perform this analysis for various seeds as well.

```python
clusters_range = range(2,15)
random_range = range(0,20)
result_kmeans =[]
for c in clusters_range:
    for r in random_range:
        clusterer = KMeans(n_clusters=c, random_state=r)
        cluster_labels = clusterer.fit_predict(cluster_scaled)
        silhouette_avg = silhouette_score(cluster_scaled, cluster_labels)
        #print("For n_clusters =", c,  "\nThe average silhouette_score is :", silhouette_avg)
        result_kmeans.append([c,r,silhouette_avg])

result_kmeans = pd.DataFrame(result_kmeans, columns=["n_clusters","seed","silhouette_score"])
pivot_km = pd.pivot_table(result_kmeans, index="n_clusters", columns="seed",values="silhouette_score")

plt.figure(figsize=(15,6))
sns.heatmap(pivot_km, annot=True, linewidths=.5, fmt='.3f', cmap=sns.cm.rocket_r)
plt.tight_layout()
```

```python
max_sh = result_kmeans['silhouette_score'].max()
min_sh = result_kmeans['silhouette_score'].min()

max_sh_clusters = result_kmeans[result_kmeans['silhouette_score'] == max_sh]
min_sh_clusters = result_kmeans[result_kmeans['silhouette_score'] == min_sh]


print("max SH", max_sh, max_sh_clusters.iloc[0])
print("min SH", min_sh, min_sh_clusters.iloc[0])
```

The heatmap above shows silhouette scores for various combinations of random state and number of clusters. The highest scores are for 2 and 3 clusters and they are relatively insensitive to seed.

I will chose 3 clusters to get more insight into data.

```python
kmeans_sel = KMeans(n_clusters=3, random_state=1).fit(cluster_scaled)
labels = pd.DataFrame(kmeans_sel.labels_)
clustered_data = cluster_data.assign(Cluster=labels)
```

```python
scatters(clustered_data, 'Cluster')
```

```python
grouped_km = clustered_data.groupby(['Cluster']).mean().round(1)
grouped_km
```

Cluster 0 – lower mean of credit amount, short duration, older customers

Cluster 1 – high mean of credit amount, long duration, middle-aged customers

Cluster 2 - lower mean of credit amount, short duration, young customers

# Clustering with Affinity Propagation

```python
preferences = np.arange(-30,-190,-10)
clusters = []

for p in preferences:
    af = AffinityPropagation(preference=p, damping=0.6, max_iter=400, verbose=False).fit(cluster_scaled)
    labels_af = pd.DataFrame(af.labels_)
    clusters.append(len(af.cluster_centers_indices_))

plt.figure(figsize=(10,7))
plt.xlabel("Preference")
plt.ylabel("Number of clusters")
plt.plot(preferences,clusters, marker='o')
```

```python
af = AffinityPropagation(preference=-140, damping=0.6, verbose=False).fit(cluster_scaled)
labels_af = pd.DataFrame(af.labels_)
n_clusters_ = len(af.cluster_centers_indices_)

clustered_data_af = cluster_data.assign(Cluster=labels_af)
scatters(clustered_data_af,'Cluster')

grouped_af = clustered_data_af.groupby(['Cluster']).mean().round(1)
```

```python
grouped_af = clustered_data_af.groupby(['Cluster']).mean().round(1)
grouped_af
```

Cluster 0 – high mean of credit amount, long duration, younger customers

Cluster 1 – low mean of credit amount, short duration, younger customers

Cluster 2 - low mean of credit amount, short duration, older customers

Cluster 3 - high mean of credit amount, middle-time duration, older customers

# KMedoids Clustering

```python
clusters_range = range(2,14)
inertias =[]

for c in clusters_range:
    kmediods = KMedoids(n_clusters=c, random_state=0).fit(cluster_scaled)
    inertias.append(kmediods.inertia_)

plt.figure()
plt.plot(clusters_range,inertias, marker='o')



clusters_range = range(2,15)
random_range = range(0,20)
result_kmediods =[]
for c in clusters_range:
    for r in random_range:
        clusterer = KMedoids(n_clusters=c, random_state=r)
        cluster_labels = clusterer.fit_predict(cluster_scaled)
        silhouette_avg = silhouette_score(cluster_scaled, cluster_labels)
        #print("For n_clusters =", c,  "\nThe average silhouette_score is :", silhouette_avg)
        result_kmediods.append([c,r,silhouette_avg])

result_kmediods = pd.DataFrame(result_kmediods, columns=["n_clusters","seed","silhouette_score"])
pivot_km = pd.pivot_table(result_kmediods, index="n_clusters", columns="seed",values="silhouette_score")

plt.figure(figsize=(15,6))
sns.heatmap(pivot_km, annot=True, linewidths=.5, fmt='.3f', cmap=sns.cm.rocket_r)
plt.tight_layout()
```

```python
kmediods_sel = KMedoids(n_clusters=6, random_state=1).fit(cluster_scaled)
labels = pd.DataFrame(kmediods_sel.labels_)
clustered_data = cluster_data.assign(Cluster=labels)
scatters(clustered_data, 'Cluster')
```

```python
max_sh = result_kmediods['silhouette_score'].max()
min_sh = result_kmediods['silhouette_score'].min()

max_sh_clusters = result_kmediods[result_kmediods['silhouette_score'] == max_sh]
min_sh_clusters = result_kmediods[result_kmediods['silhouette_score'] == min_sh]


print("max SH", max_sh, max_sh_clusters.iloc[0])
print("min SH", min_sh, min_sh_clusters.iloc[0])
```

# KMedians Clustering

```python
# Initialize centers 
initial_medians = [[0.0, 0.1, 0.2], [2.5, 0.7, 0.2]]
#Create instance of K-Medians algorithm.
kmedians_instance = kmedians(cluster_scaled, initial_medians)
# Run cluster analysis and obtain results.
kmedians_instance.process()
clusters = kmedians_instance.get_clusters()
medians = kmedians_instance.get_medians()
# Visualize clustering results.
visualizer = cluster_visualizer()
visualizer.append_clusters(clusters, cluster_scaled)
visualizer.append_cluster(initial_medians, marker='*', markersize=20)
visualizer.append_cluster(medians, marker='*', markersize=20)
visualizer.show()
```

```python
silhouette_avg = np.mean(silhouette(cluster_scaled, clusters).process().get_score()) 
print("Silhouette-Kmedians" , silhouette_avg)
```
