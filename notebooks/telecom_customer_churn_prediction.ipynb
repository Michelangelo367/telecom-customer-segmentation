The next step is to categorically encode the categorical variables of test dataset so that we can start training the model.

```python
def get_test_data():
    
    for col in df_test.select_dtypes(include=[object]):
        df_test[col] = df_test[col].astype('category').cat.codes

    y_test = df_test['Churn']

    X_test = df_test[['MonthlyRevenue', 'MonthlyMinutes',
       'TotalRecurringCharge', 'DirectorAssistedCalls', 'OverageMinutes',
       'RoamingCalls', 'PercChangeMinutes', 'PercChangeRevenues',
       'DroppedCalls', 'BlockedCalls', 'UnansweredCalls', 'CustomerCareCalls',
       'ThreewayCalls', 'ReceivedCalls', 'OutboundCalls', 'InboundCalls',
       'PeakCallsInOut', 'OffPeakCallsInOut', 'DroppedBlockedCalls',
       'CallForwardingCalls', 'CallWaitingCalls', 'MonthsInService',
       'UniqueSubs', 'ActiveSubs', 'ServiceArea', 'Handsets', 'HandsetModels',
       'CurrentEquipmentDays', 'AgeHH1', 'AgeHH2', 'ChildrenInHH',
       'HandsetRefurbished', 'HandsetWebCapable', 'TruckOwner', 'RVOwner',
       'Homeownership', 'BuysViaMailOrder', 'RespondsToMailOffers',
       'OptOutMailings', 'NonUSTravel', 'OwnsComputer', 'HasCreditCard',
       'RetentionCalls', 'RetentionOffersAccepted', 'NewCellphoneUser',
       'NotNewCellphoneUser', 'ReferralsMadeBySubscriber', 'IncomeGroup',
       'OwnsMotorcycle', 'AdjustmentsToCreditRating', 'HandsetPrice',
       'MadeCallToRetentionTeam', 'CreditRating', 'PrizmCode', 'Occupation',
       'MaritalStatus']]
        
    return X_test, y_test
```

# Model Selection and Training with different Sklearn models

```python
X_train, y_train = get_train_data()
```

```python
models = [
    LinearSVC(),
    SGDClassifier(max_iter=100, tol=1e-3), KNeighborsClassifier(),
    LogisticRegression(solver='lbfgs'), LogisticRegressionCV(cv=3),
    AdaBoostClassifier(n_estimators=100), GradientBoostingClassifier(n_estimators=100), 
    BaggingClassifier(), ExtraTreesClassifier(n_estimators=300),
    RandomForestClassifier(n_estimators=300)
]


def score_model(X, y, estimator, **kwargs):
    """
    Test various estimators.
    """
    model = Pipeline([
        ('imputer missing values', IterativeImputer()),
        ('estimator', estimator)
    ])
    
    # Instantiate the classification model and visualizer
    model.fit(X, y, **kwargs)

    actual_y  = y
    pred_y = model.predict(X)

    # Compute and return F1 (harmonic mean of precision and recall)
    print("F1-Score {}: {}".format(estimator.__class__.__name__, f1_score(actual_y, pred_y)))
    print("Precision {}: {}".format(estimator.__class__.__name__, precision_score(actual_y, pred_y)))
    print("Recall {}: {}".format(estimator.__class__.__name__, recall_score(actual_y, pred_y)))
    print("Accuracy {}: {}".format(estimator.__class__.__name__, accuracy_score(actual_y, pred_y)))
    print("\n")

for model in models:
    print("Model:", model.__class__.__name__)
    score_model(X_train, y_train, model)
```

# Evalluation of Best Models using Repeated K-Fold Cross Validation

```python
models = [
    BaggingClassifier(), ExtraTreesClassifier(n_estimators=300),
    RandomForestClassifier(n_estimators=300)
]


def score_model(X, y, estimator, **kwargs):
    """
    Test various estimators.
    """
    results = []
    model = Pipeline([
        ('imputer missing values', IterativeImputer()),
        ('estimator', estimator)
    ])
    
    # evaluate the model
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    for metrics in ["accuracy", "precision", "recall", "f1"]:
        scores = cross_val_score(model, X, y, scoring=metrics, cv=cv, n_jobs=-1)
        # store results
        results.append(scores)
        print('%s: > %.3f (%.3f)' % (metrics, mean(scores), std(scores)))
        
        
    visualizer = CVScores(model, cv=cv, scoring='f1_weighted')
    # Instantiate the classification model and visualizer
    visualizer.fit(X, y, **kwargs)
    visualizer.show()

for model in models:
    print("Model:", model.__class__.__name__)
    score_model(X_train, y_train, model)
```
